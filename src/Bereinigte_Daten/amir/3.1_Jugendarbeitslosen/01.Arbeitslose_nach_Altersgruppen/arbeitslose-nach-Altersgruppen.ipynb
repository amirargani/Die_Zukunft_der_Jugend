```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
arbeitslos = pd.read_csv('13211-0003_de.csv', delimiter=';', skiprows=5, encoding='utf-8-sig')
```

```python
# DF
arbeitslos.head()
```

```python
# Spalten löschen

cols = [
    arbeitslos.columns[i]
    for i in range(3, len(arbeitslos.columns))
    if i % 2 == 1
]

arbeitslos.drop(columns=cols, inplace=True)
```

```python
arbeitslos.head()
```

```python
arbeitslos.tail()
```

```python
# Zeilen löschen
arbeitslos = arbeitslos.drop(index = [104, 105, 106, 107])
arbeitslos.tail()
```

```python
# Unnötige Zeilen löschen
for i in range(35, 104):
    arbeitslos.drop(index=i, inplace=True)
```

```python
arbeitslos.tail(10)
```

```python
arbeitslos = arbeitslos.drop(columns=['männlich', 'weiblich', '55 Jahre und älter', 'Langzeitarbeitslose', 'Schwerbehinderte', 'Ausländer'])
arbeitslos.head()
```

```python
arbeitslos = arbeitslos.drop(columns='Unnamed: 0')
arbeitslos.columns.values[0] = 'Jahr'
arbeitslos.head()
```

```python
arbeitslos.drop(index=0, inplace=True)
arbeitslos.head()
```

```python
arbeitslos.info()
```

```python
arbeitslos['Jahr'] = pd.to_numeric(arbeitslos['Jahr']).astype(int)
arbeitslos.info()
```

```python
# Datentypen ändern
cols = ['Insgesamt', 'Jugendliche unter 20 Jahre', 'Jüngere unter 25 Jahre']
arbeitslos[cols] = arbeitslos[cols].apply(pd.to_numeric, errors='coerce').astype('Int64')

arbeitslos.info()
```

```python
arbeitslos_2010_2024 = arbeitslos[arbeitslos['Jahr'] >= 2010]
arbeitslos_2010_2024.reset_index(drop=True, inplace=True)
arbeitslos_2010_2024.head()
```

```python
# Als CSV speischern
arbeitslos_2010_2024.to_csv('arbeitlosen_nach_altergruppen_.csv', index=False, sep=';', encoding='utf-8-sig')
```
