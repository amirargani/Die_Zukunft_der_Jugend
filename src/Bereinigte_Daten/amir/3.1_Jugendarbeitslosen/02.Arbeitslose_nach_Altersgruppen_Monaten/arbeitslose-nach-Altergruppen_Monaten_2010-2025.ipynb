```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
arbeitslos = pd.read_csv('13211-0004_de_2010-2025.csv', delimiter=';', skiprows=5, encoding='utf-8-sig')
```

```python
# DF
arbeitslos.head()
```

```python
# Spalten löschen

cols = [
    arbeitslos.columns[i]
    for i in range(4, len(arbeitslos.columns))
    if i % 2 == 0
]

arbeitslos.drop(columns=cols, inplace=True)
```

```python
arbeitslos.head()
```

```python
# Zeilen löschen
arbeitslos = arbeitslos.drop(index = 0)
```

```python
arbeitslos = arbeitslos.drop(columns=['männlich', 'weiblich', '55 Jahre und älter', 'Langzeitarbeitslose', 'Schwerbehinderte', 'Ausländer'])
arbeitslos.head()
```

```python
arbeitslos = arbeitslos.drop(columns='Unnamed: 0')
arbeitslos.columns.values[0] = 'Jahr'
arbeitslos.head()
```

```python
arbeitslos.rename(columns={'Unnamed: 2': 'Alter'}, inplace=True)
```

```python
arbeitslos.info()
```

```python
arbeitslos['Jahr'] = pd.to_numeric(arbeitslos['Jahr']).astype(int)
arbeitslos.info()
```

```python
# Datentypen ändern
cols = ['Jahr', 'Insgesamt', 'Jugendliche unter 20 Jahre', 'Jüngere unter 25 Jahre']
arbeitslos[cols] = arbeitslos[cols].apply(pd.to_numeric, errors='coerce').astype('Int64')

arbeitslos.info()
```

```python
arbeitslos.columns.values[1] = 'Monat'
arbeitslos.head()
```

```python
# Als CSV speischern
arbeitslos.to_csv('arbeitlosen_mit_altergruppen_Monaten_2024_2025_.csv', index=False, sep=';', encoding='utf-8-sig')
```
