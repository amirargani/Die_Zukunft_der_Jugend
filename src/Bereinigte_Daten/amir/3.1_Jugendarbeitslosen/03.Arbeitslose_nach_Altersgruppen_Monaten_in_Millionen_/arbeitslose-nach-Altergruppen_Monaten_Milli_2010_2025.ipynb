```python
altergruppe = ['15 bis unter 25 Jahre', '25 bis unter 75 Jahre']
```

```python
len(arbeitslos.columns)
```

```python
arbeitslos_copy = arbeitslos.copy()
```

```python
arbeitslos_copy.drop(index=[0, 1], inplace=True)
arbeitslos_copy.head()
```

```python
# New DateFrame

daten = []

for _, row in arbeitslos_copy.iterrows():
    jahr = row.iloc[0]
    monat = row.iloc[1]
    for start in range(2, len(arbeitslos_copy.columns), 2):
        end = min(start + 2, len(arbeitslos_copy.columns) + 1)
        index_altergruppe = 0
        for right in range(start, end):
            dic = {
                'Jahr': jahr,
                'Monat': monat,
                'Altergruppe': altergruppe[index_altergruppe],
                'Geschlecht': "männlich" if (right == 2) or (right == 3) else "weiblich",
                'Anzahl': row.iloc[right]
                }
            if dic not in daten:
                daten.append(dic)
                index_altergruppe += 1

arbeitslos_neu_bereinigt =  pd.DataFrame(daten)

arbeitslos_neu_bereinigt
```

```python
arbeitslos_neu_bereinigt.info()
```

```python
arbeitslos_neu_bereinigt['Anzahl'] = (
    arbeitslos_neu_bereinigt['Anzahl']
    .astype(str)                # sicherstellen, dass es Strings sind
    .str.replace(',', '.', regex=False)  # Komma durch Punkt ersetzen
)
arbeitslos_neu_bereinigt
```

```python
# Datentypen ändern
cols = ['Jahr']
arbeitslos_neu_bereinigt[cols] = arbeitslos_neu_bereinigt[cols].apply(pd.to_numeric, errors='coerce').astype('Int64')

arbeitslos_neu_bereinigt.head()
```

```python
# Datentypen ändern
cols = ['Anzahl']
arbeitslos_neu_bereinigt[cols] = arbeitslos_neu_bereinigt[cols].apply(pd.to_numeric, errors='coerce').astype(float)

arbeitslos_neu_bereinigt.head()
```

```python
arbeitslos_neu_bereinigt.info()
```

```python
# Werte in Millionen konvertieren
arbeitslos_neu_bereinigt['Anzahl'] = arbeitslos_neu_bereinigt['Anzahl'] * 1000000
arbeitslos_neu_bereinigt.head()
```

```python
# Datentypen anpassen
cols = ['Anzahl']
arbeitslos_neu_bereinigt['Anzahl'] = pd.to_numeric(
    arbeitslos_neu_bereinigt['Anzahl'], errors='coerce'
).fillna(0).astype(int)

arbeitslos_neu_bereinigt.head()
```

```python
# Als CSV speischern
arbeitslos_neu_bereinigt.to_csv('arbeitlosen_nach_altergruppen_Monaten_2010_2025_Milli_.csv', index=False, sep=';', encoding='utf-8-sig')
```
