#### 1975-2024_Neu abgeschlossene_Ausbildungsstellenangebot

```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
angebot = pd.read_csv('1975_2024_Neu_abgeschlossene_Ausbildungsverträge_sowie_Ausbildungsstellenangebot_und_nachfrage_nach_Ländern.csv', delimiter=';', skiprows=0, encoding='utf-8-sig')
```

```python
# DF
angebot.head()
```

```python
angebot.drop(columns=['Neu abgeschlossene Ausbildungsverträge', 'Unnamed: 6' ,'Unnamed: 7'], inplace=True)
angebot.head()
```

```python
angebot_copy = angebot.copy()
```

```python
# Spaltennamen umbenennen
angebot_copy = angebot_copy.rename(columns={
    'Unnamed: 1': 'Jahr', 
    'Unnamed: 2': 'Insgesamt Tausend', 
    'Ausbildungsplätze Angebot / Nachfrage': 'Für das 1. Ausbildungsjahr', 
    'Unnamed: 4': 'Mit verkürzter Ausbildung', 
    'Unnamed: 5': 'Unbesetzte Ausbildungsplätze', 
    # 'Unnamed: 6': 'Unversorgte Bewerber / -innen ohne Alternative', 
    # 'Unnamed: 7': 'Bewerber / -innen mit Alternative', 
    'Unnamed: 8': 'Ausbildungsplatzangebot', 
    'Unnamed: 9': 'Ausbildungsplatznachfrage'
    })

angebot_copy.head()
```

```python
angebot_copy.drop(index=[0, 1], inplace=True)
angebot_copy.head()
```

```python
angebot_copy = angebot_copy.replace('m', 0)
angebot_copy.head()
```

```python
angebot_copy = angebot_copy.applymap(lambda x: str(x).replace(',', '.'))
angebot_copy.head()
```

```python
angebot_copy.drop(columns=['Für das 1. Ausbildungsjahr', 'Mit verkürzter Ausbildung'], inplace=True)
angebot_copy.head()
```

```python
angebot_copy.tail(50)
```

```python
len(angebot_copy)
```

```python
angebot_copy.drop(index=[2, 3, 4], inplace=True)
angebot_copy.drop(index=[i for i in range(40, 61)], inplace=True)
angebot_copy.tail(30)
```

```python
angebot_copy.drop(index=[i for i in range(5, 25)], inplace=True)
angebot_copy.head(30)
```

```python
angebot_copy.info()
```

```python
angebot_copy[['Jahr', 'Insgesamt Tausend', 'Unbesetzte Ausbildungsplätze', 'Ausbildungsplatzangebot', 'Ausbildungsplatznachfrage']] = angebot_copy[['Jahr', 'Insgesamt Tausend', 'Unbesetzte Ausbildungsplätze', 'Ausbildungsplatzangebot', 'Ausbildungsplatznachfrage']].apply(pd.to_numeric, errors='coerce').astype(float)
angebot_copy[['Jahr']] = angebot_copy[['Jahr']].apply(pd.to_numeric, errors='coerce').astype(int)
angebot_copy.info()
```

```python
# Als CSV speischern
angebot_copy.to_csv('1975_2024_Neu_abgeschlossene_Ausbildungsverträge_sowie_Ausbildungsstellenangebot_und_nachfrage_.csv', index=False, sep=';', encoding='utf-8-sig')
```
