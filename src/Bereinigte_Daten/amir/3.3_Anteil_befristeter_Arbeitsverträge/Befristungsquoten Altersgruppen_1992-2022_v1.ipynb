```python
befriste_copy = befriste.copy()
```

```python
befriste_copy.drop(index=[0, 2], inplace=True)
befriste_copy.head()
```

```python
# New DateFrame

daten = []
altergruppe = ['15-25', '25-35', '34-45', '45-55', '55-65']

for _, row in befriste_copy.iterrows():
    jahr = row.iloc[0]
    for start in range(1, len(befriste_copy.columns), 5):
        end = min(start + 5, len(befriste_copy.columns))
        index_altergruppe = 0
        for right in range(start, end):
            dic = {
                'Jahr': jahr,
                'Altersgruppe': altergruppe[index_altergruppe],
                'Geschlecht': "weiblich" if (right in range(1, 6)) else "mÃ¤nnlich",
                'Anzahl': row.iloc[right]
                }
            if dic not in daten:
                daten.append(dic)
                index_altergruppe += 1

befriste_neu_bereinigt =  pd.DataFrame(daten)

befriste_neu_bereinigt.head(10)
befriste_neu_bereinigt.tail(10)
```

```python
befriste_neu_bereinigt.info()
```
