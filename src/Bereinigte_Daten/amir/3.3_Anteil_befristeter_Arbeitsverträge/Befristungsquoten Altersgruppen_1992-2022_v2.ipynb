```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
befriste = pd.read_csv('Befristungsquoten_Altersgruppen_1992_2022.csv', delimiter=';', skiprows=4, encoding='utf-8-sig')
```

```python
# DF
befriste.head()
```

```python
befriste.drop(columns=['Unnamed: 0', 'Unnamed: 7', 'Unnamed: 13' ,'Unnamed: 14', 'Unnamed: 15' ,'Unnamed: 16', 'Unnamed: 17','Unnamed: 18' ,'Unnamed: 19','Unnamed: 20', 'Unnamed: 21'], inplace=True)
befriste.head()
```

```python
befriste.drop(index=1, inplace=True)
befriste.head()
```

```python
befriste.head(50)
```

```python
len(befriste)
```

```python
rows_delete = [
    i for i in range(35, len(befriste) + 1)
]

befriste.drop(index=rows_delete, inplace=True)

befriste.tail(20)
```

```python
befriste = befriste.applymap(lambda x: str(x).replace(",", "."))
befriste.head()
```

```python
befriste_copy = befriste.copy()
```

```python
befriste_copy.drop(index=[0, 2], inplace=True)
befriste_copy.head()
```

```python
befriste_frau = befriste_copy.drop(columns=['Unnamed: 8', 'Unnamed: 9',	'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'])
befriste_frau.head()
```

```python
# Spaltennamen umbenennen
befriste_frau = befriste_frau.rename(columns={
'Unnamed: 1': 'Jahr',
'Unnamed: 2': '15-25',
'Unnamed: 3': '25-35',
'Unnamed: 4': '35-45',
'Unnamed: 5': '45-55',
'Unnamed: 6': '55-65',
})

befriste_frau.head()
```

```python
# befriste_frau['Geschlecht'] = 'Frauen'
# befriste_frau.head()

befriste_frau.insert(
    loc=1,  # Indexposition, an der die Spalte eingefügt werden soll
    column='Geschlecht',
    value='Frauen')
```

```python
befriste_frau.head()
```

```python
befriste_frau.reset_index(drop=True, inplace=True)
befriste_frau.head()
```

```python
befriste_mann = befriste_copy.drop(columns=['Unnamed: 2', 'Unnamed: 3',	'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6'])
befriste_mann.head()
```

```python
# Spaltennamen umbenennen
befriste_mann = befriste_mann.rename(columns={
'Unnamed: 1': 'Jahr',
'Unnamed: 8': '15-25',
'Unnamed: 9': '25-35',
'Unnamed: 10': '35-45',
'Unnamed: 11': '45-55',
'Unnamed: 12': '55-65',
})

befriste_mann.head()
```

```python
befriste_mann.insert(
    loc=1,  # Indexposition, an der die Spalte eingefügt werden soll
    column='Geschlecht',
    value='Männer')

befriste_mann.head()
```

```python
befriste_neu_bereinigt =  pd.concat([befriste_frau, befriste_mann], ignore_index=True)
display(befriste_neu_bereinigt.head(50))
display(befriste_neu_bereinigt.tail(50))
```

```python
befriste_neu_bereinigt['Jahr'] = pd.to_numeric(befriste_neu_bereinigt['Jahr'], errors='coerce').astype(int)
```

```python
befriste_neu_bereinigt.info()
```

```python
befriste_neu_bereinigt[['15-25', '25-35', '35-45', '45-55', '55-65']] = befriste_neu_bereinigt[['15-25', '25-35', '35-45', '45-55', '55-65']].apply(pd.to_numeric, errors='coerce').astype(float)
befriste_neu_bereinigt.info()
```

```python
altergruppen = ['15-25', '25-35', '34-45', '45-55', '55-65']
```

```python
for col in altergruppen:
    befriste_neu_bereinigt[col] = befriste_neu_bereinigt[col].apply(lambda x: f'{str(x).replace(".", ",")}')
```

```python
# Als CSV speischern
# befriste_neu_bereinigt.to_excel('Befristungsquoten_Altersgruppen_1992_2022_v2_.xlsx', index=False)
```

```python
befriste_neu_bereinigt.to_csv('Befristungsquoten_Altersgruppen_1992_2022_v2_.csv', sep=';', index=False, encoding='utf-8-sig')
```
