```python
befriste = befriste.applymap(lambda x: str(x).replace(",", "."))
befriste.head()
```

```python
befriste_copy = befriste.copy()
```

```python
befriste_copy.drop(index=[0, 2], inplace=True)
befriste_copy.head()
```

```python
# New DateFrame

daten = []
altergruppen = ['15-25', '25-35', '34-45', '45-55', '55-65']

for _, row in befriste_copy.iterrows():
    jahr = row.iloc[0]
    
    # Weiblich: Spalten 1–6
    daten.append({
        'Jahr': jahr,
        'Geschlecht': 'weiblich',
        '15-25': row.iloc[1],
        '25-35': row.iloc[2],
        '34-45': row.iloc[3],
        '45-55': row.iloc[4],
        '55-65': row.iloc[5]
    })
    
    # Männlich: Spalten 6–11
    daten.append({
        'Jahr': jahr,
        'Geschlecht': 'männlich',
        '15-25': row.iloc[6],
        '25-35': row.iloc[7],
        '34-45': row.iloc[8],
        '45-55': row.iloc[9],
        '55-65': row.iloc[10]
    })

befriste_neu_bereinigt = pd.DataFrame(daten)
display(befriste_neu_bereinigt.head(10))
display(befriste_neu_bereinigt.tail(10))
```

```python
befriste_neu_bereinigt['Jahr'] = pd.to_numeric(befriste_neu_bereinigt['Jahr'], errors='coerce').astype(int)
```

```python
# Korrekte Altersgruppen-Spalten
altergruppen = ['15-25', '25-35', '34-45', '45-55', '55-65']
befriste_neu_bereinigt[altergruppen] = befriste_neu_bereinigt[altergruppen].apply(pd.to_numeric, errors='coerce').astype(float)
befriste_neu_bereinigt.info()
```
