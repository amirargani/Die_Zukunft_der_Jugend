Schulabgänger Deutschland

```python
# Laden der Bibliotheken 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Laden der CSV-Dateien

abgaenger = pd.read_csv("0Schulabgänger_Deutschland.csv", delimiter= ';', skiprows=4, encoding='utf-8-sig')
abgaenger
```

```python
# Spaltennamen
print(abgaenger.columns.tolist())
```

```python
# Löschen von Spalten
abgaenger = abgaenger.drop(columns=['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 7'])
abgaenger.head(6)
```

```python
print(abgaenger.columns.tolist())
```

```python
# Spaltennamen umbenennen
abgaenger = abgaenger.rename(columns={
    'Deutschland': 'Geschlecht', 
    'Unnamed: 1': 'Schuljahr', 
    'Unnamed: 2': 'Anzahl_Schüler', 
    'Unnamed: 6': 'Absolventen_und_Abgänger', 
    })

abgaenger.info()
```

```python
# Datentypen ändern 
abgaenger[['Anzahl_Schüler', 'Absolventen_und_Abgänger']] = abgaenger[['Anzahl_Schüler', 'Absolventen_und_Abgänger']].apply(pd.to_numeric, errors='coerce').astype('Int64')
```

```python
abgaenger.info()
```

```python
abgaenger.head(50)
```

#Alternativ-Code zum Umbenennen einiger Zeilen in einer bestimmten Spalte

Alternativ mit .loc und .isna kombiniert:

```python
# Umbenennen der nan-Werte zu 'männlich'

for i in range(3,29):
    if pd.isna(abgaenger.loc[i, 'Geschlecht']):
        abgaenger.loc[i, 'Geschlecht'] = 'männlich'

abgaenger.head(50)
```

```python
abgaenger.tail(50)
```

```python
# Umbenennen der nan-Werte zu 'weiblich'

for i in range(30,56):
    if pd.isna(abgaenger.loc[i, 'Geschlecht']):
        abgaenger.loc[i, 'Geschlecht'] = 'weiblich'

abgaenger.tail(50)
```

```python
abgaenger.head(50)
```

```python
# Umbenennen der nan-Werte zu 'insgesamt'

for i in range(57,83):
    if pd.isna(abgaenger.loc[i, 'Geschlecht']):
        abgaenger.loc[i, 'Geschlecht'] = 'Insgesamt'

abgaenger.tail(50)
```

```python
# Löschen von Zeilen 
abgaenger = abgaenger.drop(index =[83,84,85])
abgaenger.tail()
```

```python
abgaenger.head()
```

```python
# Löschen der ersten Zeilen 
abgaenger = abgaenger.drop(index =[0,1])
abgaenger.head()
```

```python
# Löschen von nicht bedeutsamen Zeilen für das Datafrane 
abgaenger = abgaenger.drop(index =[28, 55,82])
abgaenger.head(50)
```

```python
abgaenger.tail(50)
```

```python
abgaenger_neu = abgaenger.copy()
abgaenger_neu = abgaenger_neu[abgaenger_neu['Schuljahr'] >= '2009/10']
abgaenger_neu
```

```python
# Speichern der Datei als CSV
abgaenger_neu.to_csv('Abgaenger_allgemein.csv', sep=';', index=False, encoding='utf-8-sig')
```
