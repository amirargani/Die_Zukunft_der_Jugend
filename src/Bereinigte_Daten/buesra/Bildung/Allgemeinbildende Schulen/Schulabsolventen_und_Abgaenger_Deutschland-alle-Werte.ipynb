```python
# New DateFrame

daten = []

for _, row in abschluss_t.iterrows():
    jahr = row.iloc[0]
    if jahr in jahre:
        geschlecht = row.iloc[1]
        art = 0 # Index - Schulart-Liste
        for start in range(2, 86, 6):
            end = min(start + 6, 87)
            counter = 0 # Index - Schulabschluss-Liste
            for right in range(start, end):
                # print(start, end, right)
                dic = {
                'Schuljahr': jahr,
                'Schulart': schulart[art],
                'Schulabschluss': schulabschluss[counter],
                'Geschlecht': geschlecht,
                'Anzahl': row.iloc[right]
                }
                if dic not in daten: # oder => abschlusse_neu.drop_duplicates(inplace=True)
                    daten.append(dic)
                    counter += 1
            art +=1

abschlusse_neu =  pd.DataFrame(daten)
#abschlusse_neu.drop_duplicates(inplace=True)

# Nach ['Schuljahr', 'Schulart'] sortieren
# abschlusse_neu = abschlusse_neu.sort_values(by=['Schuljahr', 'Schulart'], ascending=[True, True])

# Schulart-Liste als kategorisch mit definierter Reihenfolge setzen
abschlusse_neu['Schulart'] = pd.Categorical(
    abschlusse_neu['Schulart'],
    categories=schulart,
    ordered=True
)

display(abschlusse_neu.head(12))
display(abschlusse_neu.tail(12))
```
