```python
# Laden der Bibliotheken 
import pandas as pd
import numpy as np
import os
```

```python
# Laden der CSV-Dateien

#  Alle .csv-Dateien im Ordner auflisten
csv_dateien = [f for f in os.listdir('Bildung\\Berufliche-Schule') if f.endswith('.csv')]

berufsschule_d = pd.DataFrame()

for f in csv_dateien:
    pfad_datei = os.path.join('Bildung\\Berufliche-Schule', f)
    berufsschule = pd.read_csv(pfad_datei, delimiter= ';', skiprows=3, encoding='latin1')
    berufsschule_d = pd.concat([berufsschule_d, berufsschule], ignore_index = True)

berufsschule_d
```

```python
# Löschen der Spalte 'Land'
berufsschule_d = berufsschule_d.drop(columns=['Land'])

# Nach Jahren sortieren
berufsschule_jahr = berufsschule_d.sort_values(by='Jahr')
berufsschule_jahr
```

```python
# NaN-Werte zu Null konvertieren
berufsschule_jahr = berufsschule_jahr.fillna(0)
berufsschule_jahr
```

```python
# # Spalten für Männlich berechnen

# Beispiel: Neue Spalte 'Geschlecht_Männlich' an Position 3 einfügen
berufsschule_jahr.insert(
    loc=3,  # Indexposition, an der die Spalte eingefügt werden soll
    column='Geschlecht_Männlich',
    value=berufsschule_jahr['Geschlecht_Insgesamt'] - berufsschule_jahr['Geschlecht_Weiblich']
)

# Weitere Spalten einfügen
berufsschule_jahr.insert(
    loc=6,
    column='Hauptabschluss_Männlich',
    value=berufsschule_jahr['Hauptabschluss_Gesamt'] - berufsschule_jahr['Hauptabschluss_Weiblich']
)

berufsschule_jahr.insert(
    loc=9,
    column='Mittlererschulabschlusss_Männlich',
    value=berufsschule_jahr['Mittlererschulabschluss_Gesamt'] - berufsschule_jahr['Mittlererschulabschluss_Weiblich']
)

berufsschule_jahr.insert(
    loc=12,
    column='Fachhochschulreife_Männlich',
    value=berufsschule_jahr['Fachhochschulreife_Gesamt'] - berufsschule_jahr['Fachhochschulreife_Weiblich']
)

berufsschule_jahr.insert(
    loc=15,
    column='Hochschulreife_Männlich',
    value=berufsschule_jahr['Hochschulreife_Gesamt'] - berufsschule_jahr['Hochschulreife_Weiblich']
)

berufsschule_jahr.insert(
    loc=18,
    column='zus_erw_schul_Teil_d_Fachhochs_reife_Männlich',
    value=berufsschule_jahr['zus_erw_schul_Teil_d_Fachhochs_reife_Gesamt'] - berufsschule_jahr['zus_erw_schul_Teil_d_Fachhochs_reife_Weiblich']
)

berufsschule_jahr
```

```python
# Jahre in einer Liste speichern
years = berufsschule_jahr['Jahr'].to_list()
years
```

```python
# Daten in der Liste überprüfen, welchen Typ die einzelnen Einträge haben
for eintrag in years:
    print(f"Wert: {eintrag}, Typ: {type(eintrag)}")
```

```python
# years Daten filtern ab Jahr 2010
years_filtered = [year for year in years if year >= 2010]
years_filtered
```

```python
# Fomatieren der Daten im Datarframe 'berufsschule_jahr' in ein neues Format

abschluss = {}

for row in berufsschule_jahr.itertuples(index=False):
    # Zeile im Dataframe 
    if row.Jahr in years_filtered:

        if row.Jahr not in abschluss: 
            # Jahr in dem Dictionary 'abschluss' speichern
            abschluss[row.Jahr] = [{'Gesamt': row.Geschlecht_Insgesamt, 'Männlich':row.Geschlecht_Männlich, 'Weiblich': row.Geschlecht_Weiblich}, 
                                   {'Hauptschulabschluss_Gesamt':row.Hauptabschluss_Gesamt, 'Hauptschulabschluss_Männlich':row.Hauptabschluss_Männlich, 'Hauptschulabschluss_Weiblich': row.Hauptabschluss_Weiblich},
                                   {'Mittlererschulabschluss_Gesamt':row.Mittlererschulabschluss_Gesamt, 'Mittlererschulabschluss_Männlich':row.Mittlererschulabschlusss_Männlich, 'Mittlererschulabschluss_Weiblich': row.Mittlererschulabschluss_Weiblich}, 
                                   {'Fachhochschulreife_Gesamt':row.Fachhochschulreife_Gesamt, 'Fachhochschulreife_Männlich':row.Fachhochschulreife_Männlich, 'Fachhochschulreife_Weiblich': row.Fachhochschulreife_Weiblich}, 
                                   {'Hochschulreife_Gesamt':row.Hochschulreife_Gesamt, 'Hochschulreife_Männlich':row.Hochschulreife_Männlich, 'Hochschulreife_Weiblich': row.Hochschulreife_Weiblich}, 
                                   {'zus_erw_schul_Teil_d_Fachhochs_reife_Gesamt':row.zus_erw_schul_Teil_d_Fachhochs_reife_Gesamt, 'zus_erw_schul_Teil_d_Fachhochs_reife_Männlich':row.zus_erw_schul_Teil_d_Fachhochs_reife_Männlich, 'zus_erw_schul_Teil_d_Fachhochs_reife_Weiblich': row.zus_erw_schul_Teil_d_Fachhochs_reife_Weiblich}]
        

print(abschluss)
```

```python
# Konvertieren der Daten in Werte -Entfernen von 'np.int64' -  Neue Struktur mit "sauberen" Werten
abschluss_clean = {}

for jahr, datensaetze in abschluss.items():
    
    # Jeder Schlüssel als Jahr im Dictionary 
    jahr_clean = int(jahr)  # Jahr in normalen int umwandeln

    # Neue Liste für jedes Key des Schlüssels
    neue_liste = []

    for eintrag in datensaetze:
        # Neues Dictionary für jedes vorhandene Dictionary im 'abschluss'-Dictionary
        neuer_eintrag = {}

        for key, value in eintrag.items():

            # Für jeden Schlüssel und Key im neuen Dictionary
            if isinstance(value, np.integer):
                neuer_eintrag[key] = int(value)
            elif pd.isna(value):
                neuer_eintrag[key] = None
            else:
                neuer_eintrag[key] = value
        neue_liste.append(neuer_eintrag)
    abschluss_clean[jahr_clean] = neue_liste

abschluss_clean
```

```python
# Neues Dataframe erstellen für das Formatieren der Daten 
# df['Spalte'] → gibt eine Serie
# df[['Spalte']] → gibt ein DataFrame


berufsschule_abschlusse = pd.DataFrame(columns=['Jahr', 'Geschlecht', 'Abschlussart', 'Anzahl'])
berufsschule_abschlusse
```

```python
# Exportieren der Daten vom Dictionary in das neue Dataframe

rows = []
for year, liste in abschluss_clean.items(): 
    #print(year, liste)
    
    for index in liste:
        # print(index)
        for eintrag, wert in index.items():

            # Informationen als Zeilen speichern, um sie in das Dataframe hinzuzufügen
            row = {
                "Jahr": year,
                "Geschlecht": eintrag,
                "Abschlussart": eintrag,
                "Anzahl": wert,
            }
            rows.append(row)

# Einfügen der Zeilen in das Dataframe
berufsschule_abschlusse = pd.DataFrame(rows)
berufsschule_abschlusse
```

```python
# Korrekturwörterbuch für fehlerhafte Schreibweisen in der Geschlechtspalte
geschlecht = {
   'Geschlecht_Insgesamt' : 'Gesamt', 
    'Geschlecht_Weiblich' : 'Weiblich', 
    'Geschlecht_Männlich' : 'Männlich', 
    'Hauptschulabschluss_Gesamt': "Gesamt", 
    'Hauptschulabschluss_Weiblich' : 'Weiblich', 
    'Hauptschulabschluss_Männlich' : 'Männlich', 
    'Mittlererschulabschluss_Gesamt': 'Gesamt',
    'Mittlererschulabschluss_Weiblich': 'Weiblich', 
    'Mittlererschulabschluss_Männlich': 'Männlich',
    'Fachhochschulreife_Gesamt': 'Gesamt',
    'Fachhochschulreife_Weiblich': 'Weiblich', 
    'Fachhochschulreife_Männlich': 'Männlich', 
    'Hochschulreife_Gesamt' : 'Gesamt', 
    'Hochschulreife_Weiblich' : 'Weiblich', 
    'Hochschulreife_Männlich' : 'Männlich', 
    'zus_erw_schul_Teil_d_Fachhochs_reife_Gesamt' : 'Gesamt', 
    'zus_erw_schul_Teil_d_Fachhochs_reife_Weiblich' : 'Weiblich',
    'zus_erw_schul_Teil_d_Fachhochs_reife_Männlich' : 'Männlich',
}

abschlussart = {
    'Gesamt' : 'Alle', 
    'Weiblich' : 'Alle', 
    'Männlich' : 'Alle', 
    'Hauptschulabschluss_Gesamt': "Hauptschulabschluss", 
    'Hauptschulabschluss_Weiblich' : 'Hauptschulabschluss', 
    'Hauptschulabschluss_Männlich' : 'Hauptschulabschluss', 
    'Mittlererschulabschluss_Gesamt': 'Mittlererschulabschluss',
    'Mittlererschulabschluss_Weiblich': 'Mittlererschulabschluss', 
    'Mittlererschulabschluss_Männlich': 'Mittlererschulabschluss',
    'Fachhochschulreife_Gesamt': 'Fachhochschulreife',
    'Fachhochschulreife_Weiblich': 'Fachhochschulreife', 
    'Fachhochschulreife_Männlich': 'Fachhochschulreife', 
    'Hochschulreife_Gesamt' : 'Hochschulreife', 
    'Hochschulreife_Weiblich' : 'Hochschulreife', 
    'Hochschulreife_Männlich' : 'Hochschulreife', 
    'zus_erw_schul_Teil_d_Fachhochs_reife_Gesamt' : 'zus-erw_Fachhochsreife', 
    'zus_erw_schul_Teil_d_Fachhochs_reife_Weiblich' : 'zus_erw_Fachhochsreife',
    'zus_erw_schul_Teil_d_Fachhochs_reife_Männlich' : 'zus_erw_Fachhochsreife',

}

# Einfügen der Korrekturen in das DataFrame
berufsschule_abschlusse = berufsschule_abschlusse.replace({"Geschlecht": geschlecht})
berufsschule_abschlusse = berufsschule_abschlusse.replace({"Abschlussart": abschlussart})

berufsschule_abschlusse
```
