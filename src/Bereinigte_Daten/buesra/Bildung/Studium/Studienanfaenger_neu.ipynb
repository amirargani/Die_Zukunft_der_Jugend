Hochschule Studienanfänger 

```python
# Laden der Bibliotheken 
import pandas as pd
```

```python
# Laden der CSV-Datei
studieanfanger = pd.read_csv("0Studienanfänger.csv", delimiter= ';', skiprows=7)
studieanfanger.head(5)
```

```python
studieanfanger.info()
```

```python
# Anzeigen der Spaltennamen

print(studieanfanger.columns.tolist())
```

```python
# Löschen von Spalten
studieanfanger = studieanfanger.drop(columns=['männlich', 'Unnamed: 2','weiblich','Unnamed: 4','Insgesamt', 'Unnamed: 6', 'männlich.1', 'Unnamed: 8','weiblich.1','Unnamed: 10','Insgesamt.1','Unnamed: 12','Unnamed: 14', 'Unnamed: 16', 'Unnamed: 18'])
studieanfanger
```

```python
# Datentypen korrigieren 
studieanfanger = studieanfanger.rename(columns={'Unnamed: 0': 'Semester', 'männlich.2': 'Männlich', 'weiblich.2': 'Weiblich', 'Insgesamt.2': 'Insgesamt'})
studieanfanger[['Männlich', 'Weiblich', 'Insgesamt']] = studieanfanger[['Männlich', 'Weiblich', 'Insgesamt']].apply(pd.to_numeric, errors='coerce').astype('Int64')
studieanfanger.head(10)
studieanfanger.info()
```

```python
studieanfanger
```

```python
studieanfanger = studieanfanger.drop(index =[27, 28, 29])
studieanfanger
```

```python
studieanfanger['Männlich (Prozent)'] = round((studieanfanger['Männlich']/studieanfanger['Insgesamt']) * 100, 2)
studieanfanger['Weiblich (Prozent)'] = round((studieanfanger['Weiblich']/studieanfanger['Insgesamt']) * 100, 2)
studieanfanger
```

```python
studieanfanger.info()
```

```python
# Filtern der Daten ab WS 2009/10
studieanfanger_jahr = studieanfanger.copy()
studieanfanger_jahr = studieanfanger_jahr[studieanfanger_jahr['Semester'] >= 'WS 2009/10']
studieanfanger_jahr
```

```python
# New DateFrame

daten = []

for _, row in studieanfanger_jahr.iterrows():
    jahr = row.iloc[0]
    
    # Männlich
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Männlich',
        'Wert': row.iloc[1],
        'Prozent': row.iloc[4],
       
    })
    
    # Weiblich
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Weiblich',
        'Wert': row.iloc[2],
        'Prozent': row.iloc[5],
       
    })
    
     # Insgesamt
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Insgesamt',
        'Wert': row.iloc[3],
        'Prozent': 0,
       
    })
   

studieanfanger_neu = pd.DataFrame(daten)

display(studieanfanger_neu.head(10))
display(studieanfanger_neu.tail(10))
```

```python
# Speichern der Datei als CSV
studieanfanger_neu.to_csv('Studienanfaenger_neu.csv', sep=';', index=False, encoding='utf-8-sig')
```
