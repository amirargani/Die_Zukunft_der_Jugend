Hochschule Studienanfänger 

```python
# Laden der Bibliotheken 
import pandas as pd
```

```python
# Laden der CSV-Datei
studierende = pd.read_csv("1Studierende.csv", delimiter= ';', skiprows=7)
studierende.head(5)
```

```python
studierende.info()
```

```python
# Anzeigen der Spaltennamen

print(studierende.columns.tolist())
```

```python
# Löschen von Spalten
studierende = studierende.drop(columns=['männlich', 'Unnamed: 2','weiblich','Unnamed: 4','Insgesamt', 'Unnamed: 6', 'männlich.1', 'Unnamed: 8','weiblich.1','Unnamed: 10','Insgesamt.1','Unnamed: 12','Unnamed: 14', 'Unnamed: 16', 'Unnamed: 18'])
studierende
```

```python
# Datentypen korrigieren 
studierende = studierende.rename(columns={'Unnamed: 0': 'Semester', 'männlich.2': 'Männlich', 'weiblich.2': 'Weiblich', 'Insgesamt.2': 'Insgesamt'})
studierende[['Männlich', 'Weiblich', 'Insgesamt']] = studierende[['Männlich', 'Weiblich', 'Insgesamt']].apply(pd.to_numeric, errors='coerce').astype('Int64')
studierende.info()
```

```python
studierende
```

```python
studierende = studierende.drop(index =[27, 28, 29])
studierende
```

```python
studierende['Männlich (Prozent)'] = round((studierende['Männlich']/studierende['Insgesamt']) * 100, 2)
studierende['Weiblich (Prozent)'] = round((studierende['Weiblich']/studierende['Insgesamt']) * 100, 2)
studierende
```

```python
studierende_jahr = studierende.copy()
studierende_jahr = studierende_jahr[studierende_jahr['Semester'] >= 'WS 2009/10']
studierende_jahr
```

```python
# New DateFrame

daten = []

for _, row in studierende_jahr.iterrows():
    jahr = row.iloc[0]
    
    # Männlich
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Männlich',
        'Wert': row.iloc[1],
        'Prozent': row.iloc[4],
       
    })
    
    # Weiblich
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Weiblich',
        'Wert': row.iloc[2],
        'Prozent': row.iloc[5],
       
    })
    
     # Insgesamt
    daten.append({
        'Semester': jahr,
        'Geschlecht': 'Insgesamt',
        'Wert': row.iloc[3],
        'Prozent': 0,
       
    })
   

studierende_neu = pd.DataFrame(daten)

display(studierende_neu.head(10))
display(studierende_neu.tail(10))
```

```python
# Speichern der Datei als CSV
studierende_neu.to_csv('Studierende_neu.csv', sep=';', index=False, encoding='utf-8-sig')
```
