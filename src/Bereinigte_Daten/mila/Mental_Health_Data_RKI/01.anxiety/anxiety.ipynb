```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
anxiety=pd.read_excel(r"Anxiety_Symptoms_Age_Group_Exc.xlsx")
```

```python
# DF
anxiety.head(50)                
```

```python
# Spalten löschen
cols = ['Unnamed: 0','Unnamed: 1','Unnamed: 2']

anxiety.drop(columns=cols, inplace=True) 
anxiety.head()
```

```python
anxiety.loc[anxiety.index.isin(range(2, 5)) & anxiety['Unnamed: 3'].isna(), 'Unnamed: 3'] = '18-29 years'
anxiety.loc[anxiety.index.isin(range(6, 9)) & anxiety['Unnamed: 3'].isna(), 'Unnamed: 3'] = '30-44 years'
anxiety.loc[anxiety.index.isin(range(10, 13)) & anxiety['Unnamed: 3'].isna(), 'Unnamed: 3'] = '45-64 years'
anxiety.loc[anxiety.index.isin(range(14, 17)) & anxiety['Unnamed: 3'].isna(), 'Unnamed: 3'] = '65+ years'
anxiety
```

```python
anxiety.drop(index= [1,5,9,13], inplace=True)
anxiety
```

```python
anxiety_copy=anxiety.copy()
anxiety_t = anxiety_copy.set_index(anxiety_copy.columns[0]).T
anxiety_t
```

```python
anxiety_t.columns
```

```python
anxiety_t.reset_index(drop=True, inplace=True)
anxiety_t
```

```python
anxiety_t.columns.values[0]= 'Datum'
anxiety_t
```

```python
# New DateFrame

daten = []

for _, row in anxiety_t.iterrows():
    jahr = row.iloc[0]
    
    # Fit
    daten.append({
        'Datum': jahr,
        'Wert': 'FIT',
        '18-29': row.iloc[1],
        '30-44': row.iloc[4],
        '45-64': row.iloc[7],
        '+65':   row.iloc[10],
       
    })
    
    # Maximum FIT 
    daten.append({
        'Datum': jahr,
        'Wert': 'Maximum Fit',
        '18-29': row.iloc[2],
        '30-44': row.iloc[5],
        '45-64': row.iloc[8],
        '+65':   row.iloc[11],
       
    })
    
     # Minimum Fit
    daten.append({
        'Datum': jahr,
        'Wert': 'Minimum FIT',
        '18-29': row.iloc[3],
        '30-44': row.iloc[6],
        '45-64': row.iloc[9],
        '+65':   row.iloc[12],
       
    })
   

anxiety_neu = pd.DataFrame(daten)
anxiety_neu.drop(index=[0,1,2], inplace= True)
display(anxiety_neu.head(10))
display(anxiety_neu.tail(10))
```
