```python
# Bibliotheke
import pandas as pd
```

```python
# CSV
depressive =pd.read_excel(r"Depressive_Symptome_Age_Group_Exc.xlsx")
```

```python
# DF
depressive.head(50)                
```

```python
# Spalten löschen
cols = ['Unnamed: 0','Unnamed: 1','Unnamed: 2']

depressive.drop(columns=cols, inplace=True) 
depressive.head()
```

```python
depressive.loc[depressive.index.isin(range(2, 5)) & depressive['Unnamed: 3'].isna(), 'Unnamed: 3'] = '18-29 years'
depressive.loc[depressive.index.isin(range(6, 9)) & depressive['Unnamed: 3'].isna(), 'Unnamed: 3'] = '30-44 years'
depressive.loc[depressive.index.isin(range(10, 13)) & depressive['Unnamed: 3'].isna(), 'Unnamed: 3'] = '45-64 years'
depressive.loc[depressive.index.isin(range(14, 17)) & depressive['Unnamed: 3'].isna(), 'Unnamed: 3'] = '65+ years'
depressive
```

```python
depressive.drop(index= [1,5,9,13], inplace=True)
depressive
```

```python
depressive_copy=depressive.copy()
depressive_t = depressive_copy.set_index(depressive_copy.columns[0]).T
depressive_t
```

```python
depressive_t.columns
```

```python
depressive_t.reset_index(drop=True, inplace=True)
depressive_t
```

```python
depressive_t.columns.values[0]= 'Datum'
depressive_t
```

```python
# New DateFrame

daten = []

for _, row in depressive_t.iterrows():
    jahr = row.iloc[0]
    
    # Fit
    daten.append({
        'Datum': jahr,
        'Wert': 'FIT',
        '18-29': row.iloc[1],
        '30-44': row.iloc[4],
        '45-64': row.iloc[7],
        '+65':   row.iloc[10],
       
    })
    
    # Maximum FIT 
    daten.append({
        'Datum': jahr,
        'Wert': 'Maximum Fit',
        '18-29': row.iloc[2],
        '30-44': row.iloc[5],
        '45-64': row.iloc[8],
        '+65':   row.iloc[11],
       
    })
    
     # Minimum Fit
    daten.append({
        'Datum': jahr,
        'Wert': 'Minimum FIT',
        '18-29': row.iloc[3],
        '30-44': row.iloc[6],
        '45-64': row.iloc[9],
        '+65':   row.iloc[12],
       
    })
   

depressive_neu = pd.DataFrame(daten)
depressive_neu.drop(index=[0,1,2], inplace= True)
display(depressive_neu.head(10))
display(depressive_neu.tail(10))
```
